* Dropping ID column because it will not give us meaningful values for this analysis.

```python
reason_columns = pd.get_dummies(df['Reason for Absence'])
reason_columns
```

```python
reason_columns['check'] = reason_columns.sum(axis = 1)
reason_columns
```

```python
reason_columns['check'].nunique()
```

```python
reason_columns = reason_columns.drop(['check'], axis = 1)
reason_columns
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
reason_columns
```

```python
df = df.drop(['Reason for Absence'], axis = 1)
df.head()
```

```python
reason_type1 = reason_columns.loc[:, 1:14].max(axis = 1)
reason_type2 = reason_columns.loc[:, 15:17].max(axis = 1)
reason_type3 = reason_columns.loc[:, 18:21].max(axis = 1)
reason_type4 = reason_columns.loc[:, 22:].max(axis = 1)
```

```python
reason_type2
```

```python
df = pd.concat([df, reason_type1, reason_type2, reason_type3, reason_type4], axis = 1)
df.head()
```

```python
df.columns.values
```

```python
columns_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns = columns_names
df.head()
```

```python
columns_name_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4' ,'Date',
                          'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[columns_name_reordered]
df.head()
```

```python
df_reason_mod = df.copy()
```

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
```

```python
df_reason_mod['Date']
```

```python
df_reason_mod['Date'][0]
```

```python
df_reason_mod.info()
```

```python
list_months = []
```

```python
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
list_months
```

```python
df_reason_mod['Month Value'] = list_months
df_reason_mod.head()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod.head()
```

```python
df_reason_mod = df_reason_mod.drop(['Date'], axis = 1)
df_reason_mod.head()
```

```python
df_reason_mod.columns.values
```

```python
column_names_upd = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value',
       'Day of the week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df_reason_mod = df_reason_mod[column_names_upd]
```

```python
df_reason_date_mod = df_reason_mod.copy()
```
